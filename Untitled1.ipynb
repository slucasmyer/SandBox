```python
df = pd.read_csv('../../Desktop/pe-19.csv')[4:]
df.columns = df.iloc[0]
df = df[1:]
states = sorted(list(set(df['State Name'])))
years = sorted(list(set(df['Year of Estimate'])))
pops = {}
for state in states:
    pops[state] = {}
    for year in years:
        pops[state][year] = 0
        st_yr_df = df.loc[df['Year of Estimate'] == year].loc[df['State Name'] == state]
        st_yr_df = st_yr_df.loc[:, 'Under 5 years':]
        for i in range(len(st_yr_df)):
            for j in list(st_yr_df.columns):
                pops[state][year] += float(re.sub('\D', '', st_yr_df.iloc[i][j]))
```
